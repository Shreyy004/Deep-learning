```python
#SIGMOID 
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid_function(x):
    return 1 / (1 + np.exp(-x))

# Define the derivative of the sigmoid function
def sig_der(x):
    return sigmoid_function(x) * (1 - sigmoid_function(x))

# Generate random inputs
X = np.linspace(-10, 10, 100)
y = sigmoid_function(X)
z = sig_der(X)

# Plot the sigmoid function
plt.figure(figsize=(10, 6))
plt.plot(X, y, label='Sigmoid Function', color='red')
plt.title('Sigmoid Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()

# Plot the derivative of the sigmoid function
plt.figure(figsize=(10, 6))
plt.plot(X, z, label='Derivative of Sigmoid Function', color='blue')
plt.title('Derivative of Sigmoid Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Well-suited for binary classification tasks.(Often used in the final layer of neural networks)
# The sigmoid function forms the basis of logistic regression, mapping input values to probabilities.

# σ(x)= 1/1+e ^−x
 
# range->[0,1]

# σ′(x)=σ(x)⋅(1−σ(x))

# Derivative/Differential of the sigmoid function f’(x) will lie between 0 and 0.25.

# S-shaped Curve: The sigmoid function has an "S"-shaped curve, with:

# σ(x)→0 as x→−∞
# σ(x)→1 as x→+∞
# σ(0)=0.5

# The sigmoid activation function suffers from 
# vanishing gradients (For very large or very small input values, the gradient of the sigmoid becomes very small, leading to slow learning in deep networks), non-zero centricity ( The outputs are always positive), and slow 
# learning due to output saturation for extreme 
# input values

# Pros of the Sigmoid Function:
# Outputs values in a bounded range, making it easy to interpret as probabilities.
# Well-suited for binary classification tasks.
# Cons of the Sigmoid Function:
# Vanishing Gradient Problem: For very large or very small input values, the gradient of the sigmoid becomes very small, leading to slow learning in deep networks.
# Output Not Zero-Centered: The outputs are always positive, which can slow convergence when used with gradient descent.
# Computationally Expensive: It involves exponential calculations, which can be slower compared to simpler activation functions.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sinc function
def sinc_function(x):
    return np.sinc(x / np.pi)  # numpy's sinc function is normalized to pi

# Define the derivative of the sinc function
def sinc_der(x):
    return np.cos(x) / np.pi - np.sinc(x / np.pi) * x

# Generate random inputs
X = np.linspace(-10, 10, 100)
y = sinc_function(X)
z = sinc_der(X)

# Plot the sinc function
plt.figure(figsize=(10, 6))
plt.plot(X, y, label='Sinc Function', color='red')
plt.title('Sinc Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()

# Plot the derivative of the sinc function
plt.figure(figsize=(10, 6))
plt.plot(X, z, label='Derivative of Sinc Function', color='blue')
plt.title('Derivative of Sinc Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```

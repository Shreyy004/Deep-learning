```python
#TAN
import numpy as np
import matplotlib.pyplot as plt

# Tanh function and its derivative
def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x) ** 2

# Generate input data
X = np.linspace(-10, 10, 100)
y = tanh(X)
z = tanh_derivative(X)

# Plot tanh function
plt.figure(figsize=(3, 1))
plt.plot(X, y, label='Tanh Function', color='red')
plt.title('Tanh Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()

# Plot derivative of tanh
plt.figure(figsize=(3, 1))
plt.plot(X, z, label='Derivative of Tanh Function', color='blue')
plt.title('Derivative of Tanh Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```

```python
Pros of Tanh:
Zero-Centered Output:
helps gradients flow better during backpropagation.
It avoids the skewed updates problem caused by the non-zero centricity of sigmoid.
Smooth and Differentiable:
The function is smooth, making it suitable for gradient-based optimization.

Cons of Tanh:
Vanishing Gradients:
this leads to the vanishing gradient problem, similar to sigmoid.                                          
tanh(x)= (e^x - e ^−x)/ (e^x + e ^−x)
Range: Outputs values between −1 and 1, making it zero-centered.
tanh ′(x)=1−(tanh(x)) 
```

```python
import numpy as np
```

```python
def perceptronModel(inputs, weights, bias):
    activation = np.dot(inputs, weights) + bias
    return 0 if activation <= 0 else 1
```

```python
def updation(training_data, learning_rate, epochs):
    w = np.random.randn(len(training_data[0][0]))
    b = 0                                  
    for epoch in range(epochs):
        print(f"\nEpoch {epoch + 1}:")
        for x, target in training_data:
            y = perceptronModel(x, w, b)  
            error = target - y            
            
            w += learning_rate * error * x
            b += learning_rate * error
        
        print(f"Weights: {w}, Bias: {b}")
    
    return w, b
```

```python
training_data = [(np.array([0, 0]), 0),(np.array([0, 1]), 0),
                (np.array([1, 0]), 0),(np.array([1, 1]), 1)]
```

```python
learning_rate = 0.1
epochs = 100
weights, bias = updation(training_data, learning_rate, epochs)

print("\nTesting trained perceptron:")
test_cases = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
for test in test_cases:
    output = perceptronModel(test, weights, bias)
    print(f"AND({test[0]}, {test[1]}) = {output}")
```
